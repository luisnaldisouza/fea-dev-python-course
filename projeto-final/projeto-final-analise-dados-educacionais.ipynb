## **📊 Descrição das Colunas do DataFrame**

### **Identificação do Estudante**  
- **ID_Aluno**: Identificador único para cada aluno (de 1001 a 3392).

### **👥 Detalhes Demográficos**  
- **Idade**: Idade dos estudantes, variando de 15 a 18 anos.  
- **Genero**: Gênero do estudante:
  - 0: Masculino
  - 1: Feminino  
- **Etnia**: Etnia dos estudantes:
  - 0: Caucasiano
  - 1: Afrodescendente
  - 2: Asiático
  - 3: Outra

### **👨‍👩‍👧 Educação dos Pais**  
- **Escolaridade_Pais**: Nível de escolaridade dos pais:
  - 0: Nenhum
  - 1: Ensino Médio
  - 2: Alguma Faculdade
  - 3: Graduação
  - 4: Pós-Graduação ou Superior

### **📚 Hábitos de Estudo**  
- **Horas_Estudo_Semanais**: Tempo de estudo semanal em horas (0 a 20).  
- **Faltas**: Número de faltas no ano letivo (0 a 30).  
- **Aulas_Particulares**: Participação em aulas particulares:
  - 0: Não
  - 1: Sim

### **👨‍👩‍👧 Envolvimento Parental**  
- **Apoio_Pais**: Nível de apoio dos pais:
  - 0: Nenhum
  - 1: Baixo
  - 2: Moderado
  - 3: Alto
  - 4: Muito Alto

### **🏫 Atividades Extracurriculares**  
- **Atividades_Extracurriculares**: Participação em atividades extracurriculares:
  - 0: Não
  - 1: Sim  
- **Esportes**: Participação em esportes:
  - 0: Não
  - 1: Sim  
- **Musica**: Participação em atividades musicais:
  - 0: Não
  - 1: Sim  
- **Voluntariado**: Participação em atividades de voluntariado:
  - 0: Não
  - 1: Sim

### **📈 Desempenho Acadêmico**  
- **Media_Notas**: Média das notas (GPA), variando de 2.0 a 4.0.  
- **Classificacao_Notas**: Classificação do aluno com base na média das notas:
  - 0: 'A' (Média >= 3.5)
  - 1: 'B' (3.0 <= Média < 3.5)
  - 2: 'C' (2.5 <= Média < 3.0)
  - 3: 'D' (2.0 <= Média < 2.5)

## **Importação e exploração dos dados**

```python
import pandas as pd

# Carregar o arquivo CSV
df = pd.read_csv("dados_estudantes.csv", index_col=False)
df
```

```python
df["GradeClass"] = df["GPA"].apply(lambda x: "A" if x >= 3.5 else "B" if x >= 3 else "C" if x >= 2.5 else "D" if x >= 2 else "F")

df.head(100)
```

```python

# Renomear colunas
df = df.rename(columns={
    "StudentID": "ID_Aluno",                
    "Age": "Idade",
    "Gender": "Gênero",
    "Ethnicity": "Etnia",
    "ParentalEducation": "Escolaridade_Pais",
    "StudyTimeWeekly": "Horas_Estudo_Semanais",
    "Absences": "Faltas",
    "Tutoring": "Aulas_Particulares",
    "ParentalSupport": "Apoio_Pais",
    "Extracurricular": "Atividades_Extracurriculares",
    "Sports": "Esportes",
    "Music": "Musica",
    "Volunteering": "Voluntariado",
    "GPA": "Media_Notas",
    "GradeClass": "Classificacao_Notas"
})

# Substituir valores nas colunas conforme solicitado
df['Gênero'] = df['Gênero'].replace({0: 'Masculino', 1: 'Feminino'})
df['Etnia'] = df['Etnia'].replace({0: 'Caucasiano', 1: 'Afrodescendente', 2: 'Asiático', 3: 'Outra'})
df['Escolaridade_Pais'] = df['Escolaridade_Pais'].replace({0: 'Nenhum', 1: 'Ensino Médio', 2: 'Alguma Faculdade', 3: 'Graduação', 4: 'Pós-Graduação ou Superior'})
df['Apoio_Pais'] = df['Apoio_Pais'].replace({0: 'Nenhum', 1: 'Baixo', 2: 'Moderado', 3: 'Alto', 4: 'Muito Alto'})
df['Atividades_Extracurriculares'] = df['Atividades_Extracurriculares'].replace({0: 'Não', 1: 'Sim'})
df['Esportes'] = df['Esportes'].replace({0: 'Não', 1: 'Sim'})
df['Musica'] = df['Musica'].replace({0: 'Não', 1: 'Sim'})
df['Aulas_Particulares'] = df['Aulas_Particulares'].replace({0: 'Não', 1: 'Sim'})
df['Voluntariado'] = df['Voluntariado'].replace({0: 'Não', 1: 'Sim'})


df
```

```python
# Fazer uma coluna qua vai ordenar o DataFrame a partir dos alunos com melhores valores em Classificacao_Notas e Media_Notas fazendo uma coluna Ranking e assim ranjiando os alunos também.
df_sorted = df.sort_values(by=["Classificacao_Notas"], ascending=[True])

# Criar a coluna Ranking
df_sorted["Ranking"] = range(1, len(df_sorted) + 1)

# Reordenar colunas para que "Ranking" seja a primeira coluna
columns_order = ["Ranking"] + [col for col in df_sorted.columns if col != "Ranking"]
df = df_sorted[columns_order].reset_index(drop=True)

df
```

## **Análise exploratória e exploração de dados**

 **Pergunta: Atividades extracurriculares, fazer um esporte, ou fazer aula de musica diminui ou aumenta os valores de Classificao_Notas e Media_Notas?**

```python
extra_notas_media = df.groupby('Atividades_Extracurriculares')['Media_Notas'].mean()
musica_notas_media = df.groupby('Musica')['Media_Notas'].mean()
esportes_notas_media = df.groupby('Esportes')['Media_Notas'].mean()

fig = plt.figure(figsize=(15, 6))

# Gráfico para Média de Notas
plt.bar(['Extra Não', 'Extra Sim'], extra_notas_media.values, color=['skyblue', 'lightgreen'], label='Extra')
plt.bar(['Música Não', 'Música Sim'], musica_notas_media.values, color=['skyblue', 'lightgreen'], label='Música', alpha=0.6)
plt.bar(['Esporte Não', 'Esporte Sim'], esportes_notas_media.values, color=['skyblue', 'lightgreen'], label='Esporte', alpha=0.4)
plt.title('Comparação de Média de Notas')
plt.xlabel('Atividade')
plt.ylabel('Média de Notas')
plt.legend(loc='upper left')



plt.tight_layout()
plt.show()
```

Análise: Alunos que fazem uma extracurricular, esporte e aula de musica tem um considerável aumento na Média_notas, indicando que fazer atividades fora as atividades acadêmicas pode aumentar seu desempenho na mesma.

 **Pergunta: Atividades extracurriculares, fazer um esporte, ou fazer aula de musica diminui ou aumenta o tempo de estudo semanal dos alunos?**:

```python
# Calcular médias de horas de estudo semanal para cada categoria (esportes, Musica e extracurriculares)
extracurricular_media = df.groupby('Atividades_Extracurriculares')['Horas_Estudo_Semanais'].mean()
esports_media = df.groupby('Esportes')['Horas_Estudo_Semanais'].mean()
musica_media = df.groupby('Musica')['Horas_Estudo_Semanais'].mean()

print("Média de Horas de Estudo Semanal:")
print("\nAtividades Extracurriculares:")
print(extracurricular_media)
print("\nEsportes:")
print(esports_media)
print("\nMúsica:")
print(musica_media)
```

```python
fig = plt.figure(figsize=(18, 6))

# Gráfico para Atividades Extracurriculares
plt.subplot(1, 3, 1)
plt.bar(['Não', 'Sim'], extracurricular_media.values,  color=['skyblue', 'lightgreen'])
plt.title('Atividades Extracurriculares')
plt.xlabel('Participação')
plt.ylabel('Média de Horas de Estudo Semanal')

# Gráfico para Esportes
plt.subplot(1, 3, 2)
plt.bar(['Não', 'Sim'], esports_media.values, color=['skyblue', 'lightgreen'])
plt.title('Esportes')
plt.xlabel('Participação')

# Gráfico para Música
plt.subplot(1, 3, 3)
plt.bar(['Não', 'Sim'], musica_media.values, color=['skyblue', 'lightgreen'])
plt.title('Música')
plt.xlabel('Participação')

plt.tight_layout()
plt.show()
```

Análise: Não há aumento nem diminuição visível que altere algo nas horas semanais de estudo dos alunos com base nas atividades extracurriculares, esporte e musica.

 **Pergunta: A quantidade de faltas altera o comprometimento e desempenho acadêmico dos alunos?**:

```python
faltas_media = df.groupby('Faltas')['Media_Notas'].mean()


plt.figure(figsize=(7, 6))

# Gráfico 2: Faltas x Media_Notas
plt.plot(faltas_media.index, faltas_media.values, marker='o', color='blue')
plt.title("Impacto do Número de Faltas na Média de Notas")
plt.xlabel("Número de Faltas")
plt.ylabel("Média de Notas")
plt.grid(True, linestyle='--', alpha=0.5)


plt.tight_layout()
plt.show()
```

Analise: Impressionantemente quanto mais faltas pior o desempenho dos valores da Média_Notas, como era óbvio, quanto mais o aluno não vai as aulas, pior o desempenho academico dele.

 **Pergunta: Se os pais possuem uma alta escolaridade e dão o apoio devido ao aluno ele possui um melhor desempnho acadêmico?**:

```python
# Filtrar alunos com alta escolaridade dos pais e alto apoio parental
filtro_alta_escolaridade_apoio = df[(df['Escolaridade_Pais'].isin(['Graduação', 'Pós-Graduação ou Superior'])) &
                                   (df['Apoio_Pais'].isin(['Alto', 'Muito Alto']))]

# Calcular a média de notas desses alunos
media_alta_escolaridade_apoio = filtro_alta_escolaridade_apoio['Media_Notas'].mean()

# Calcular a média de notas dos demais alunos
filtro_demais = df[~((df['Escolaridade_Pais'].isin(['Graduação', 'Pós-Graduação ou Superior'])) &
                     (df['Apoio_Pais'].isin(['Alto', 'Muito Alto'])))]
media_demais = filtro_demais['Media_Notas'].mean()

# Dados para o gráfico
categorias = ['Alta Escolaridade + Alto Apoio', 'Demais Alunos']
medias = [media_alta_escolaridade_apoio, media_demais]

# Criar o gráfico de barras
plt.figure(figsize=(8, 6))
cores = ['blue', 'gray']  # Diferenciar as cores das barras
plt.bar(categorias, medias, color=cores)

# Configurar o gráfico
plt.title('Comparação de Média de Notas', fontsize=16)
plt.xlabel('Categoria', fontsize=14)
plt.ylabel('Média de Notas', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Adicionar os valores das médias acima das barras
for i, v in enumerate(medias):
    plt.text(i, v + 0.02, f"{v:.2f}", ha='center', fontsize=12)

# Exibir o gráfico
plt.tight_layout()
plt.show()
```

Análise: Sim, como pode ver no gráfico acima, a tendência é que quanto mais escolaridade dos pais e maior o apoio que eles dão ao aluno, maior é o desempenho escolar do mesmo.
